```python
sys.path.append(os.path.abspath('.'))
input_path = os.path.join('..', 'data', 'town_list.csv')
df = pd.read_csv(input_path)

```

```python

def add_coordinates(df, town_col='town_key'):
    """
    Adds latitude and longitude columns to a DataFrame based on town and state information.
    
    Parameters:
    - df (pd.DataFrame): DataFrame containing columns for town and state.
    - town_col (str): The name of the column with town names.
    
    Returns:
    - pd.DataFrame: The original DataFrame with two new columns: 'latitude' and 'longitude'.
    """
    # Initialize the geocoder with a user agent string
    geolocator = Nominatim(user_agent="geoapiExercises")
    
    # Lists to store latitude and longitude values
    latitudes = []
    longitudes = []
    
    # Iterate through the DataFrame rows
    for idx, row in df.iterrows():
        # Combine town and state to form a query string
        query = row[town_col]
        try:
            location = geolocator.geocode(query)
            if location:
                latitudes.append(location.latitude)
                longitudes.append(location.longitude)
            else:
                latitudes.append(None)
                longitudes.append(None)
        except Exception as e:
            print(f"Error geocoding '{query}': {e}")
            latitudes.append(None)
            longitudes.append(None)
        # Pause to respect Nominatim's usage policy
        time.sleep(1)
    
    # Create new columns in the DataFrame
    df['latitude'] = latitudes
    df['longitude'] = longitudes
    return df
```
